```python
class TiendaOnline: 

    def __init__(self):
        self.inventario = [{'nombre': 'patata', 'precio': 2.5, 'cantidad': 60}, {'nombre': 'manzana', 'precio': 2, 'cantidad': 35}, {'nombre': 'lechuga', 'precio': 3, 'cantidad': 70}] 
        self.ventas_totales = 0.0 
    
    def eliminar_producto(self, nombre):
        for p in self.inventario:
            if p["nombre"] == nombre:
                self.inventario.remove(p)
                return 
        print("El producto no está en el inventario")
```

```python
class TiendaOnline: 

    def __init__(self):
        self.inventario = [{'nombre': 'patata', 'precio': 2.5, 'cantidad': 60}, {'nombre': 'manzana', 'precio': 2, 'cantidad': 35}] 
        self.ventas_totales = 0.0 
    
    def calcular_valor_inventario(self):
        total_inventario = 0
        for v in self.inventario:
            valor_producto = v['precio'] * v['cantidad']
            total_inventario += valor_producto
        print(f"El valor total del inventario es: {total_inventario}€")
```
