```python
# MEJORAS EN EL CÓDIGO

class TiendaOnline: 

    def __init__(self):
        self.inventario = [{'nombre': "patata", 'precio': 2, 'cantidad': 2}, {'nombre': "manzana", 'precio': 3, 'cantidad': 3}] 
        self.ventas_totales = 0.0 
    
    def agregar_producto(self, nombre, precio, cantidad): 
        for p in self.inventario:
            if p["nombre"] == nombre: # Búsqueda directa dentro de una lista
                p["cantidad"] += cantidad # Búsqueda directa dentro de una lista
                return        
        producto_nuevo = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}  # Manera simplificada de crear el diccionario
        self.inventario.append(producto_nuevo) 
    
    def ver_inventario(self): 
        for i in self.inventario: 
            print(f"Nombre: {i['nombre']}, Precio: {i['precio']}€, Cantidad: {i['cantidad']}")
    
    def buscar_producto(self, nombre): # Sale mejor como el apartado de arriba 
        for p in self.inventario:
            if p.get("nombre") == nombre: # .get(k) -> Usamos .get() para obtener el nombre de forma segura (key) en un diccionario
                print(p)
    
    def actualizar_stock(self, nombre, cantidad):
        for s in self.inventario:
            if s.get("nombre") == nombre: # .get(k) -> Usamos .get() para obtener el nombre de forma segura (key) en un dicionario
                s["cantidad"] += cantidad # Igual a .get(k), pero directo en un diccionario
                return 
        print("El producto no está en el inventario")
    
    def eliminar_producto(self, nombre):
        for p in self.inventario:
            if p.get("nombre") == nombre: # # # .get(k) -> Usamos .get() para obtener el nombre de forma segura (key) en un dicionario
                indice = self.inventario.index(p) # .index() -> Usamos .index() para encontrar el índice donde se posiciona el nombre y sus características (el diccionario) del producto en una lista
                self.inventario.pop(indice) # .pop() -> Usamos .pop() para eliminar el producto de la lista (Buscamos su índice y lo extraemos)
                return 
        print("El producto no está en el inventario")

    def calcular_valor_inventario(self):
        total_inventario = 0.0
        for v in self.inventario:
            valor_producto = v['precio'] * v['cantidad'] # Búsqueda directa en un dicccionario
            total_inventario += valor_producto
        print(f"El valor total del inventario es: {total_inventario}€")
```
