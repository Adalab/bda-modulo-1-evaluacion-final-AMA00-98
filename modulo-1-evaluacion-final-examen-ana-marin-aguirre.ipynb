```python
# Ejercicio IMPORTANTE --> Class TiendaOnline 

class TiendaOnline: 

    def __init__(self):
        self.inventario = [] 
        self.ventas_totales = 0.0 
    
    def agregar_producto(self, nombre, precio, cantidad): 
        for p in self.inventario:
            if p["nombre"] == nombre:
                p["cantidad"] += cantidad #
                return #       
        producto_nuevo = {}
        producto_nuevo.update({"nombre": nombre, "precio": precio, "cantidad": cantidad})
        self.inventario.append(producto_nuevo) #
    
    def ver_inventario(self): 
        for i in self.inventario:
            print(f"nombre: {i['nombre']}, precio: {i['precio']}, cantidad: {i['cantidad']}")
    
    def buscar_producto(self, nombre):
        for p in self.inventario:
            if p["nombre"] == nombre:
                print(f"nombre: {p['nombre']}, precio: {p['precio']}, cantidad: {p['cantidad']}") #
    
    def actualizar_stock(self, nombre, cantidad):
        for s in self.inventario:
            if s["nombre"] == nombre:
                s["cantidad"] += cantidad #
                return #
        print("El producto no est치 en el inventario")
    
    def eliminar_producto(self, nombre):
        for p in self.inventario:
            if p["nombre"] == nombre:
                self.inventario.remove(p)
                return #
        print("El producto no est치 en el inventario")
    
    def calcular_valor_inventario(self):
        total_inventario = 0
        for v in self.inventario:
            valor_producto = v['precio'] * v['cantidad']
            total_inventario += valor_producto
        self.ventas_totales = total_inventario
        return self.ventas_totales #
```
